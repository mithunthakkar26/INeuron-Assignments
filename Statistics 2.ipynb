#### Problem Statement 4: <br>
Based on population figures and other general information on the U.S. population, <br>
suppose it has been estimated that, on average, a family of four in the U.S. spends <br>
about $1135 annually on dental expenditures. Suppose further that a regional dental <br>
association wants to test to determine if this figure is accurate for their area of <br>
country. To test this, 22 families of 4 are randomly selected from the population in <br>
that area of the country and a log is kept of the family’s dental expenditure for one <br>
year. The resulting data are given below. Assuming, that dental expenditure is <br>
normally distributed in the population, use the data and an alpha of 0.5 to test the <br>
dental association’s hypothesis. <br>
1008, 812, 1117, 1323, 1308, 1415, 831, 1021, 1287, 851, 930, 730, 699, <br>
872, 913, 944, 954, 987, 1695, 995, 1003, 994

```python

mu = 1135
alpha = 0.05/2
n = 22
x = ss.mean([1008, 812, 1117, 1323, 1308, 1415, 831, 1021, 1287, 851, 930, 730, 699, 872, 913, 944, 954, 987, 1695, 995, 1003, 994])
s = ss.stdev([1008, 812, 1117, 1323, 1308, 1415, 831, 1021, 1287, 851, 930, 730, 699, 872, 913, 944, 954, 987, 1695, 995, 1003, 994])
#H0: a family of four in the U.S. spends about $1135 on average annually on dental expenditures
#H1: a family of four in the US does not spend $1135 on average annually on dental expenditures
df = n-1

Test_Stat = (x-mu)/(s/np.sqrt(n))
print(Test_Stat)

print(f"P value is {t.cdf(Test_Stat, df)} at a significance level of {alpha} since this is a 2-tailed test. Hence we failed to reject the null and conclude that a family of four in the U.S. spends about $1135 on average annually on dental expenditures")
```

The manufacturers of compact disk players want to test whether a small price <br>

#### Problem Statement 15: <br> 
Comparisons of two population proportions when the hypothesized difference is zero <br>
Carry out a two-tailed test of the equality of banks’ share of the car loan market in <br>
1980 and 1995. <br>
Population 1: 1980 <br>
n1 = 1000 <br>
x1 = 53 <br>
𝑝 1 = 0.53 <br>
Population 2: 1985 <br>
n2 = 100 <br>
x2 = 43 <br>
𝑝 2= 0.43 <br>

```python
n1 = 1000
x1 = 53
𝑝1 = 0.53

n2 = 100
x2 = 43
𝑝2= 0.43

alpha = 0.05

#H0: Banks' share of car loan market in 1980 and 1985 is equal
#H1: Banks' share of car loan market in 1980 and 1985 is not equal

count = np.array([x1, x2])
nobs = np.array([n1, n2])
stat, pval = proportions_ztest(count, nobs,alternative='two-sided')
print(f"Pvalue of the test is {pval} which is lower than the significance level of {alpha}. Hence, we reject the null hypothesis and conclude that Banks' share of car loan market in 1980 and 1985 is not equal")



```

#### Problem Statement 17: <br>
A die is thrown 132 times with the following results: Number turned up: 1, 2, 3, 4, 5, 6 <br>
Frequency: 16, 20, 25, 14, 29, 28 <br>
Is the die unbiased? Consider the degrees of freedom as 𝑝 −1 . <br>

```python
Obs = [16,20,25,14,29,28]
#Every throw of a die has 1/6 probability of success. Hence expected successes of all 132 throws would be 132/6 = 22 
Exp = [22,22,22,22,22,22]
alpha = 0.05

#H0: The die is unbiased
#H1: The die is biased
print(f"PValue of {chisquare(Obs, f_exp=Exp)[1]} is higher than the significance level of {alpha}. Hence we fail to reject the null hypothesis and conclude that the die is unbiased.")
```

#### Problem Statement 20: <br>
Children of three ages are asked to indicate their preference for three photographs of <br>
adults. Do the data suggest that there is a significant relationship between age and <br>
photograph preference? What is wrong with this study? [Chi-Square = 29.6, with 4 <br>
df: 𝑝 < 0.05]. <br>




```python
data = {'5-6 years':[18, 22, 20], '7-8 years':[2, 28, 40], '9-10 years': [20,10,40]}
df = pd.DataFrame(data, index = ["A", "B", "C"]) 
alpha = 0.05
#H0: there is no significant relationship between age and photograph preference
#H1: there is a significant relationship between age and photograph preference
print(1 - stats.chi2.cdf(29.6, 4))
df=df.transpose()
print(f"PValue of {stats.chi2_contingency(df)[1]} is lower than significance level of {alpha}. Hence we reject the null hypothesis and conclude that there is a significant relationship between age and photograph preference.")
```
